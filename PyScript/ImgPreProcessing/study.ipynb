```python
import os

img_li = []

for root, dirs, files in os.walk("./ThemeImg_Prepro"):
    for f in files:
        if f.endswith(".png"):
            img_li.append(os.path.join(root, f))
```

```python
from sklearn.cluster import KMeans
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img_li[0]
```

```python
img = cv2.imread(img_li[0])
imghsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


plt.figure(figsize=(10,5))
plt.imshow(imghsv)
```
